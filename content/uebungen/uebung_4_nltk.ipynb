# Übung: NLP mit NLTK
Abgabe: **21. Dezember**

Du vereinst in dieser Übung viele Inhalte des Kurses. Fülle die Codezellen aus und beantworte die Textfragen direkt darunter.

## Aufgabe 1: Text laden
Lade einen Text aus `nltk.corpus.gutenberg` (z. B. *Austen Emma*). Zeige die ersten 20 Tokens.

## Aufgabe 2: Tokenisierung
Tokenisiere den Text mit `word_tokenize` und gib die Anzahl Tokens aus.

## Aufgabe 3: Satzzeichen entfernen
Bereinige die Tokens um Satzzeichen (regex erlaubt). Wie viele Tokens bleiben übrig?

## Aufgabe 4: Stoppwörter
Entferne englische Stoppwörter. Wie ändert sich die Länge der Tokenliste? Schreibe eine kurze Reflexion.

## Aufgabe 5: Frequenzanalyse
Erstelle eine `FreqDist` und zeige die 20 häufigsten Wörter.

## Aufgabe 6: Visualisierung
Plotte die Verteilung der 30 häufigsten Wörter (Balkendiagramm).

## Aufgabe 7: Type-Token-Ratio
Berechne die TTR und interpretiere das Resultat in 2–3 Sätzen.

## Aufgabe 8: Eigene Funktion
Schreibe eine Funktion `clean_text(tokens)` die Satzzeichen entfernt, Stoppwörter löscht und alles lowercase setzt. Teste sie.

## Aufgabe 9: Fehlerbehandlung
Erweitere deine Funktion aus Aufgabe 8 mit `try/except`. Sie soll reagieren wenn kein Iterable übergeben wird.

## Aufgabe 10: Mini-Analyse
Formuliere in 3–5 Sätzen was die häufigsten Wörter über den Text aussagen. Was überrascht dich?
