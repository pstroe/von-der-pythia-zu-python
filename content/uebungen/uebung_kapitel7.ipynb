## Aufgabe 1: Deine erste Klasse
Erstelle eine Klasse `Coin`, die folgende Attribute im Konstruktor (`__init__`) setzt:
- `issuer` (String)
- `weight` (Float)
- `mint_year` (Integer)

Schreibe ausserdem eine Methode `info()`, die alle Attribute schön formatiert ausgibt (z. B. per f-String).

```python
# Schreibe deinen Code hier

class Coin:
    def __init__(self, issuer, weight, mint_year):
        self.§ = issuer
        self.weight = weight
        self.mint_year = mint_year

    def info(self):
        return f"{self.issuer},\n{self.weight},\n{self.mint_year}"

    def wear(self, amount):
        self.wear_amount = amount
        self.weight -= amount
```
